```python


def combinedstring(inputf, start_pattern, stop_pattern):
    global recording
    for line in open(inputf, encoding="utf-8").readlines():
        if recording is False:
            if re.search(start_pattern, re.sub(' +',' ',line.encode('ascii', 'ignore').decode('ascii'))) is not None:
                recording = True
                output_section.append(re.sub(' +',' ',line.encode('ascii', 'ignore').decode('ascii')).strip())
        elif recording is True:
            if re.search(stop_pattern, re.sub(' +',' ',line.encode('ascii', 'ignore').decode('ascii'))) is not None:
                recording = False
            output_section.append(re.sub(' +',' ',line.encode('ascii', 'ignore').decode('ascii')).strip())
    return ''.join(map(str, output_section))
        
     
```

```python
from collections import Counter
from nltk.stem import WordNetLemmatizer
def posive(inputposve):
    
    tokens1 = word_tokenize(inputposve)
    
    # Convert the tokens into lowercase: lower_tokens
    lower_tokens1 = [t.lower() for t in tokens1]

# Create a Counter with the lowercase tokens: bow_simple
    bow_simple1 = Counter(lower_tokens1)

# Print the 10 most common tokens
        #print(bow_simple1.most_common(10))

# Retain alphabetic words: alpha_only
    alpha_only1 = [t for t in lower_tokens1 if t.isalpha()]


# Remove all stop words: no_stops
    no_stops1 = [t for t in alpha_only1 if t not in english_stops]


# Instantiate the WordNetLemmatizer
    wordnet_lemmatizer = WordNetLemmatizer()

# Lemmatize all tokens into a new list: lemmatized
    lemmatized1 = [wordnet_lemmatizer.lemmatize(t) for t in no_stops1]
    average_sentence_length = len(lemmatized1)/len(sent_tokenize(inputposve))
# Create the bag-of-words: bow
    bow1 = Counter(lemmatized1)
# Print the 10 most common tokens
        #print(bow1.most_common(10))
    return lemmatized1, average_sentence_length
```
